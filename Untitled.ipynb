# Regression Model

```python
# necessary packages
import numpy as np
import pandas as pd

from scipy import stats
import statsmodels.api as sm

import matplotlib.pyplot as plt
```

## Scenario: 
### Y: Housing price (price) 
### X: Living area (sqft_living)
Here we use the dataset download from:
https://www.kaggle.com/datasets/sukhmandeepsinghbrar/housing-price-dataset?resource=download

```python
data = pd.read_csv('/Users/jiangyanze/Downloads/Housing.csv')
```

```python
Y = data['price']
X = data['sqft_living']
```

```python
X
```

```python
Y
```

```python
plt.scatter(X,Y)
plt.axis([0,9000,50000,6000000])
plt.xlabel('price')
plt.ylabel('living area')
plt.show()
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y,X1).fit()
```

```python
reg.summary()
```
