# Financial Frontier

```python
# necessary packages
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
```

## Two-assets Portfolio

```python
assets = ['PG','^GSPC']
data = pd.DataFrame()

for stock in assets:
    data[stock] = yf.download(stock, start = '2014-01-01', end = '2023-12-31')['Close']
```

```python
data
```

```python
( data / data.iloc[0] * 100).plot(figsize = (12, 12))
```

```python
log_return = np.log(data / data.shift(1))
log_return.mean()*250
log_return.cov()*250
log_return.corr()
```

```python
num_assets = len(assets)
```

```python
# to generate random weights
weights = np.random.random(num_assets)
weights /= np.sum(weights)
weights
```

### Expected Portfolio Return

```python
np.sum(weights * log_return.mean() * 250)
```

### Expected Portfolio ReturnÂ¶

```python
np.dot(weights.T,np.dot(weights * log_return.cov() * 250, weights ))
```

### Expected Portfolio Volatility

```python
np.sqrt(np.dot(weights.T,np.dot(weights * log_return.cov() * 250, weights )))
```

### Simulate 1000 portfolio

```python
pfolio_return = []
pfolio_volati = []

for x in range(1000):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    pfolio_return.append(np.sum(weights * log_return.mean() * 250))
    pfolio_volati.append(np.sqrt(np.dot(weights.T,np.dot(weights * log_return.cov() * 250, weights ))))

pfolio_return = np.array(pfolio_return)
pfolio_volati = np.array(pfolio_volati)

pfolio_return, pfolio_volati
```

#### Construct Portfolio based on simulations 

```python
pfolio = pd.DataFrame( { 'Return' : pfolio_return, 'Volatility' : pfolio_volati } ) 
pfolio
```

```python
pfolio.plot( x = 'Volatility', y = 'Return', kind = 'scatter', figsize = (12, 12) )
plt.xlabel('Expected Volatility')
plt.ylabel('Expected Return')
```
