```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/jiangyanze/Downloads/Housing.csv')
```

```python
data
```

```python
X = data[["bedrooms","bathrooms","sqft_living"]]
Y = data["price"]
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y,X1).fit()

reg.summary()
```
